```python
import pylab as pl
import numpy as np
import scipy as sp
from scipy import stats
import seaborn 
seaborn.set_style('darkgrid')
```

```python
np.random.seed(456)
mymean = 100 #media che scelgo di usare per tutte le distribuzioni che andrò a creare
```

```python
df= mymean #setto il numero di gradi di libertà uguale alla media, chi2!!
p=0.5
n=mymean / p #mean=p*n #binomial

#voglio creare 100 samples, quindi ho bisogno di una stuttura che li contiene 

mydist={} #così ho creato un dizionario vuoto 

pl.hist(np.random.chisquare(mymean, size=500));
pl.ylabel('N')
pl.xlabel('x');
```

```python
mysize = (np.random.rand(100) * 2000).astype(int)
#np.random.rand(100) mi dà 100 numeri a caso fra 0 e 1
#(np.random.rand(100) * 2000) così moltiplico tutti i numeri creati per 2000
mysize[mysize <10]=10
pl.plot(mysize,'.') 
```

```python
for s in mysize:
  mydist[s]=np.random.chisquare(mymean, size=s)

ax=pl.figure().add_subplot(111)

for d in mydist.keys():
  ax.plot(d,mydist[d].mean(),'.')

ax.plot(pl.xlim(),[100,100], 'k', alpha=0.5)
ax.set_xlabel('sample size')
ax.set_ylabel('sample mean')
```
